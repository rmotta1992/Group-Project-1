```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
Women_Health = "Fem_Reproductive_Health.csv"
Women_Health_df = pd.read_csv(Women_Health)
Women_Health_df
```

```python
Women_Health_Clean_df = Women_Health_df.drop(Women_Health_df.index[[0,1]])
Women_Health_Clean_df
```

```python
Women_Health_Clean_df.columns.to_list()
```

```python
#I need to create a DataFrame that will show me units sold so I can plot the graph

Fem_Health_Units = [ x for x in Women_Health_Clean_df if x.startswith ("Units") == True]

Fem_Health_Units_Filtered = Women_Health_Clean_df.loc[:,
    ['Total US',
     'BC DEPARTMENT',
     'BC CATEGORY',
     'BC SUB CATEGORY',
     'BRAND OWNER HIGH',
     'BRAND HIGH',
     'BRAND FAMILY',
     'SHORT PRODUCT DESCRIPTION'] + Fem_Health_Units]
Fem_Health_Units_Filtered

#ppe_data_df.columns.to_list()
#[list of column names] + nd

#ppe_data_df.loc[:,]
# + nd]
#nd = [x for x in ppe_data_df.columns if x.startswith('Numerical Distribution')==True]

# new data frame =
```

```python
#dataframe for graph
monthly_units = pd.concat([pd.to_numeric(Fem_Health_Units_Filtered[col], errors='coerce') for col in Fem_Health_Units_Filtered], axis=1).sum()

monthly_units_df = pd.DataFrame(monthly_units) 
monthly_units_df

monthly_units_df1 = monthly_units_df.drop(['Total US',
     'BC DEPARTMENT',
     'BC CATEGORY',
     'BC SUB CATEGORY',
     'BRAND OWNER HIGH',
     'BRAND HIGH',
     'BRAND FAMILY',
     'SHORT PRODUCT DESCRIPTION'])

monthly_units_df1

monthly_units_df2 = monthly_units_df1.reset_index(drop=False)
monthly_units_df2

monthly_units_df3 = monthly_units_df2.rename(columns= {"index": "Month",
                                                     0: "Total Units Sold"})

monthly_units_df3
```

```python
fig = plt.figure(dpi = 200)
#bar graph for units sold monthly_units_df3
x_value1 = monthly_units_df3["Month"]
y_value1 = monthly_units_df3["Total Units Sold"]

monthly_units_bar = plt.bar(x_value1, y_value1)
plt.xticks(x_value1, rotation="vertical")
plt.xlabel("Month")
plt.ylabel("Total Units Sold")
plt.title("# Female Reproductive Health Products Sold By Month")
plt.tick_params(axis='x', which='major', labelsize=5)
plt.show()
#monthly_units_bar
```
