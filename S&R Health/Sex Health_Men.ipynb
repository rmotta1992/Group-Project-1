```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
Condoms_Data = "Condoms_Data.csv"
Condoms_Data_df = pd.read_csv(Condoms_Data)
Condoms_Data_df
```

```python
Male_Health_Clean_df = Condoms_Data_df.drop(Condoms_Data_df.index[[0,1]])

Male_Health_Clean_df
```

```python
Male_Health_Clean_df.columns.to_list()


```

```python
#I need to create a DataFrame that will show me units sold so I can plot the graph

Male_Health_Units = [ x for x in Male_Health_Clean_df if x.startswith ("Units") == True]

Male_Health_Units_Filtered = Male_Health_Clean_df.loc[:,
    ['Total US',
     'BC DEPARTMENT',
     'BC CATEGORY',
     'BC SUB CATEGORY',
     'BRAND OWNER HIGH',
     'BRAND HIGH',
     'BRAND FAMILY',
     'SHORT PRODUCT DESCRIPTION'] + Male_Health_Units]

Male_Health_Units_Filtered
```

```python
#dataframe for graph
monthly_units = pd.concat([pd.to_numeric(Male_Health_Units_Filtered[col], errors='coerce') for col in Male_Health_Units_Filtered], axis=1).sum()

monthly_units_df = pd.DataFrame(monthly_units) 
monthly_units_df
```

```python
monthly_units_df1 = monthly_units_df.drop(['Total US',
     'BC DEPARTMENT',
     'BC CATEGORY',
     'BC SUB CATEGORY',
     'BRAND OWNER HIGH',
     'BRAND HIGH',
     'BRAND FAMILY',
     'SHORT PRODUCT DESCRIPTION'])

monthly_units_df1
```

```python
monthly_units_df2 = monthly_units_df1.reset_index(drop=False)
monthly_units_df2
```

```python
monthly_units_df3 = monthly_units_df2.rename(columns= {"index": "Month",
                                                     0: "Total Units Sold"})

monthly_units_df3
```

```python
fig = plt.figure(dpi = 200)
#bar graph for units sold monthly_units_df3
x_value1 = monthly_units_df3["Month"]
y_value1 = monthly_units_df3["Total Units Sold"]

monthly_units_bar = plt.bar(x_value1, y_value1)
plt.xticks(x_value1, rotation="vertical")
plt.xlabel("Month")
plt.ylabel("Total Units Sold")
plt.title("# Male Reproductive Health Products Sold By Month")
plt.tick_params(axis='x', which='major', labelsize=5)

plt.show()
```
